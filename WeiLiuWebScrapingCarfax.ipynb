```python
df = pd.DataFrame()
for file in filenames:
    sub_ = pd.read_csv(file)
    df = df.append(sub_, ignore_index=True)
df =df.dropna(axis=0)
df.index = np.arange(0,len(df))
df

```

```python
for i in np.arange(len(df)):
    df.mileage[i] = int(re.split(' ', df.mileage[i])[1].replace(',',''))
    df.price[i] = int(df.price[i].replace('$','').replace(',',''))
df
```

```python
 df['specs'].map(lambda x: drivecheck(x))
```

```python
def drivecheck(x):
    if 'FWD' in x:
        return 'FWD'
    elif 'RWD' in x:
        return 'RWD'
    elif 'AWD' in x:
        return 'AWD'
    else:
        return 'NotSpecified'
df['drive'] = df['specs'].map(lambda x:drivecheck(x))
df
```

```python
re.split()
```

```python
df['year'] = df['model'].map(lambda x: re.split(' ', x)[0])
df['modelname'] = df['model'].map(lambda x: ' '.join(re.split(' ', x)[2:]))
df['make'] = df['model'].map(lambda x: re.split(' ', x)[1])
df.head()
#print(type(df.modelname))
```
